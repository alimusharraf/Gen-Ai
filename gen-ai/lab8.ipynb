```python
#Lab-8

import cohere
import getpass
from string import Template

# Load the Text File
file_path = "Teaching.txt"

try:
    with open(file_path, "r", encoding="utf-8") as file:
        text_content = file.read()
        print("File loaded successfully!")
except Exception as e:
    print("Error loading file:", str(e))

# Enter API key from keyboard
COHERE_API_KEY = getpass.getpass("Enter your Cohere API Key: ")

# Initialize Cohere client
co = cohere.Client(COHERE_API_KEY)

# Prompt template
template = Template("""
You are an AI assistant helping to summarize and analyze a text document.

Here is the document content:
$text

Summary:
Provide a concise summary of the document.
""")

formatted_prompt = template.substitute(text=text_content)

# Generate response
response = co.chat(
    model="command-r-08-2024",
    message=formatted_prompt
)

print("\n **Formatted Output** ")
print(response.text)
```
